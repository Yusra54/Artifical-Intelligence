```python
# Repeat the above activity to find the smallest and largest element of the list. (Suppose all the elements are integer values) 

list1=[]
list2=[]
list3=[]
print("---------Enter values for list 1----------")
for i in range(2):
  val=int(input("Enter a number: "))
  list1.append(val)
print("---------Enter values for list 2----------")
for i in range(2):
  val=int(input("Enter a number: "))
  list2.append(val)

list3=list1+list2

print(list3)
max=0
min=0
length=len(list3)-1

for i in range(0,length,1):
  if list3[i]<list3[i+1]:
    max=list3[i+1]
    min=list3[i]
print("Maximum value is: ",max)
print("Minimum value is: ",min)

```

```python
# Create a dictionary by extracting the keys from a given dictionary 
# Write a Python program to create a new dictionary by extracting the mentioned keys from the below dictionary. 
 
# Given dictionary: 
 
# sample_dict = {     "name": "Kelly", 
#    "age": 25, 
#    "salary": 8000, 
#    "city": "New york"} 
 
# Keys to extract keys = ["name", "salary"] 

sample_dict = {"name": "Kelly", "age": 25, "salary": 8000, "city": "New york"} 
keys = ["name", "salary"]
new_dict={}
for key in keys:
  new_dict[key] = sample_dict[key]
print(new_dict)

```
